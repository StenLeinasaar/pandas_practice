## Basic Plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("./data/olympics.csv", skiprows=4)
```

```python
data.head()
```

### Plot Types

**What were the different sports in the first olympics? Plot them using different graphs**

```python
first_olympics = data[data.Edition == 1896]
first_olympics.head()
```

### **Line plot**

```python
first_olympics.Sport.value_counts().plot(kind='line');
```

```python
first_olympics.Sport.value_counts().plot(kind='bar');
```

```python
first_olympics.Sport.value_counts().plot(kind='barh');
```

```python
first_olympics.Sport.value_counts().plot(kind='pie');
```

## Plot Colors

specify the name of the color specifing the color property. Matplot website has a list of built in colors

```python
first_olympics.Sport.value_counts().plot(color="plum");
```

```python
first_olympics.Sport.value_counts().plot(kind='bar', color="maroon");
```

Figure size is a tuple where we can specify width and height

```python
first_olympics.Sport.value_counts().plot(color="maroon", figsize=(10,3));
```

## Colormaps

Finding a color scheme for a data. Three classes: sequential, diverging, qualitative. Sequential is when there is order. Diverging when information deviates aroundmiddle value usally two different colors. Qualitative has no ordering

```python
first_olympics.Sport.value_counts().plot(kind='pie', colormap="Pastel1");
```

## Seaborn basic plotting 

Attractive statistical plots, a complement to matplot, not substitute. 

countplot, hue is categorixal variables, data is the dataframe or source of data. Order paramater is for sequence. Palette is for colors of different hue levels

```python
import seaborn as sns
```

### **How Many mendals have been won by men and women in the history of the olympics. How many gold, silve, and bronze medals were won for each gender?**

```python
sns.countplot(x='Medal', data=data, hue = 'Gender')
```

### Plotting challenge

#### **Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using matplotlib and seaborn.**

Using matplotlib 

```python
chinese_team = data[(data.NOC == "CHN") & (data.Edition == 2008)]
```

```python
chinese_team.Medal.value_counts().plot(kind='bar');
```

using seaborn

```python
sns.countplot(x='Medal', data=chinese_team);
```

```python
sns.countplot(x='Medal', data = chinese_team, hue = "Gender", palette="bwr");
```

#### **How can you use colormaps to give the data more meaning?**

To emphasise some attribute like above I did with gender. It makes it more intuitive to notice gender difference in medal winning. It is interesting that women earn more gold medals, silver medals and bronze medals than men.
