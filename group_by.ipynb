Group by splits a dataframe into groups based on some criteria.
Applies a function to each group independently. 
Combines the results into a dataframe

```python
data.groupby("Edition")
```

```python
type(data.groupby("Edition"))
```

```python
list(data.groupby("Edition"))
```

Iterate through a group

```python
for group_key, group_value in data.groupby('Edition'):
    print(group_key)
    print(group_value)
```

```python
type(group_value)
```

```python
data.groupby('Edition').size()
```

```python
data.groupby(['Edition', 'NOC', 'Medal']).agg(['min', 'max', 'count'])
```

```python
data.groupby(['Edition', 'NOC', 'Medal']).size()
```

```python
data.groupby(['Edition', 'NOC', 'Medal']).agg({"Edition": ['min', 'max', 'count']})
```

#### **Using groupby, plot the total number of medals awarded at each of the olympic games throughout history**

```python
data.groupby('Edition').agg({"Medal": 'count'}).plot();
```

```python
data.groupby("Edition").size().plot();
```

## Stack and unstack 

```python
mw = data[(data.Edition == 2008) & ( (data.Event == '100m') | (data.Event == "200m"))]
```

```python
mw.head()
```

```python
g = mw.groupby(["NOC", "Gender", "Discipline","Event"]).size()
g
```

```python
df = g.unstack(["Discipline", "Event"])
df
```

Stack returns a dataframe or series. It drops all NA values

```python
df.stack()
```

```python
df.stack("Event")
```

Unstack returns dataframe. In a time of Nan value we can specify the value 

```python
df
```

```python
df.unstack()
```
