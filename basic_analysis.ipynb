### Boolean indexing

boolean vectors can be used to filter data. Multiple conditions must be grouped using brackets

```python
data.head()
```

```python
data[data.Medal == 'Gold']
```

```python
data[data.NOC == "EST"]
```

```python
data[(data.Gender == 'Women') & (data.Medal == "Gold")]
```
